```python
users = pd.read_csv('sandy_tweets_user_temp_1.csv', index_col=0)
users.head()
```

## SQL

```python
%load_ext sql
```

```python
%sql DROP TABLE IF EXISTS users
%sql PERSIST users
%sql SELECT * FROM users LIMIT 5;
```

```python
%%sql sqlite://
DROP VIEW IF EXISTS users_mindate;
CREATE VIEW  users_mindate AS
--For each user_id find min time_stamp
SELECT user_id, followers_count, MIN(time_stamp) FROM users
GROUP BY user_id;
SELECT * FROM users_mindate LIMIT 5 --Print first 5 rows
```

```python
%%sql sqlite://
--Select only the  for (camis, maxinspdate) combination. I.e. filter out all the other inspdates
DROP VIEW IF EXISTS users_filt;
CREATE VIEW  users_filt AS
SELECT a.user_id, a.followers_count, a.screen_name, a.time_stamp, b.followers_count, a.followers_count*1.0/b.followers_count ratio
  FROM users a --a is alias
   JOIN users_mindate b
   ON a.user_id = b.user_id;
SELECT * FROM users_filt LIMIT 5
```

```python
%%sql sqlite://
SELECT * FROM users_filt 
WHERE followers_count>10000
ORDER BY ratio DESC LIMIT 100;
```
