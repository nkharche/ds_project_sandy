### Feature extraction

```python
# Define stop-words
from nltk.corpus import stopwords
import string
 
punctuation = list(string.punctuation)
stop = stopwords.words('english') + punctuation
```

```python
# Count word (monogram) frequencies. Remove stop-words
# https://marcobonzanini.com/2015/03/17/mining-twitter-data-with-python-part-3-term-frequencies/

from collections import Counter
import operator 

def remove_shortwords(text, stop, min_length):
    text_wo_shortwords = [term.lower() for term in text.split() if len(term)>min_length]# and len(term)>3]
    return " ".join(text_wo_shortwords)

def remove_stopwords(text, stop):
    text_wo_stopwords = [term.lower() for term in text.split() if term.lower() not in stop]# and len(term)>3]
    return " ".join(text_wo_stopwords)



#df['text_wo_stopwords'] = df['text'][df['pow_out_label']==1].apply(remove_stopwords, args=(stop,))
df['text_wo_shortwords'] = df['text'].apply(remove_shortwords, args=(stop,3))
df['text_wo_stopwords'] = df['text_wo_shortwords'].apply(remove_stopwords, args=(stop,))
#df.head()

#monogram_tokens = Counter(" ".join(df['text_wo_stopwords'].dropna()).split()).most_common(20)
monogram_tokens = Counter(" ".join(df['text_wo_stopwords']).split()).most_common(100)

df.drop(['text_wo_shortwords', 'text_wo_stopwords'],inplace=True,axis=1)

df_monogram_tokens = pd.DataFrame(monogram_tokens, columns=['monograms', 'count'])
df_monogram_tokens#.head()
```

### N-grams, N=1,2,3

```python
monograms = ['blackout', 'dark', 'darker', 'nopower', 'powerless']
bigrams = ['black out', 'power down', 'power off', 'power out', 'lose power', 'losing power', 
           'lost power', 'no power', 'power outage', 'goodbye power', 'taken power', 
           'w/o power', 'transformer exploding', 'transformer explosion','without power', 
           'without pow',
           'power back']
trigrams = ['downed electrical wires',  'power not expected', 
            'knock out power', 'knocked out power',
            'off the grid', 'shut off power', 'wait power return'] 
ngrams = monograms + bigrams + trigrams
```

```python
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))
stopwords_keep = ['no', 'the', 'out', 'off', 'not']
[stop.discard(w) for w in stopwords_keep]
```

```python
import re

def stopword_filtered_word_feats(text, words_all):
    l = re.split('[\s"]+',string.strip(text))

    #print (words_all)
    feat = dict([(word, False) for word in words_all])
    feat.update((k, True) for k, v in feat.items() if k in text.lower())#.split())
    return feat
```

```python
ngram_feats = df[['text', 'pow_out_label']].apply(lambda x: stopword_filtered_word_feats(x[0], ngrams), axis=1).tolist()
ngram_feats[:2]
```

## Sentiment analysis

```python
from textblob import TextBlob
```

```python
def calc_sentiment(text):
    # pass tweet into TextBlob
    tweet = TextBlob(text.decode("utf-8", "replace"))
    
    # output sentiment polarity
    if (tweet.sentiment.polarity<0):
        return 1 # Negative
    elif (tweet.sentiment.polarity==0):
        return 2 # Neutral
    else:
        return 3 # Positive
```

```python
tweet_sentiment = df['text'].apply(calc_sentiment).values
df['sentiment'] = tweet_sentiment
```

## Power outage classifier

```python
tweet_feats = []
for i,feat in enumerate(ngram_feats):
    #feat['sentiment'] = tweet_sentiment[i]
    #feat['num_words'] = num_words[i]
    tweet_feats.append(feat)
```

```python
tweet_feats[:1]
```

```python
# A function to evaluate our classifier

import collections
import nltk.classify.util, nltk.metrics
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import movie_reviews
 
def evaluate_classifier(feats, labels):    
    
    full_set = zip(feats, labels)
    
    posfeats = [full_set[i] for i,l in enumerate(labels) if l==1]
    negfeats = [full_set[i] for i,l in enumerate(labels) if l==0]    

    negcutoff = int(len(negfeats)*3/4)
    poscutoff = int(len(posfeats)*3/4)
 
    trainfeats = negfeats[:negcutoff] + posfeats[:poscutoff]
    testfeats = negfeats[negcutoff:] + posfeats[poscutoff:]
     
    classifier = NaiveBayesClassifier.train(trainfeats)
    refsets = collections.defaultdict(set)
    testsets = collections.defaultdict(set)
    
    for i, (feats, label) in enumerate(testfeats):
            refsets[label].add(i)
            observed = classifier.classify(feats)
            testsets[observed].add(i)
     
    print ('accuracy:', nltk.classify.util.accuracy(classifier, testfeats))
    print ('pos precision:', nltk.precision(refsets[1], testsets[1]))
    print ('pos recall:', nltk.recall(refsets[1], testsets[1]))
    print ('neg precision:', nltk.precision(refsets[0], testsets[0]))
    print ('neg recall:', nltk.recall(refsets[0], testsets[0]))
    classifier.show_most_informative_features(50)
    
    return classifier
```

```python
# Accuracy using baseline bag of words, without stop words
classifier = evaluate_classifier(tweet_feats, df['pow_out_label'].values)
```

```python
import dill
dill.dump(classifier, open('power_outage_NB_classifier.dill', 'w'))
```
